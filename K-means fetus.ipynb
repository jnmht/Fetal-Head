```python
import glob
import numpy as np
from PIL import Image
import os, numpy, PIL
import matplotlib.pyplot as plt
%matplotlib inline
import cv2
import json
from os.path import exists

train_data_files = sorted(glob.glob("data/train/images/*.png"))
print('Train image files found ', len(train_data_files))

train_annotations_files = sorted(glob.glob("data/train/annotations/*.png"))
print('Train annotation files found ', len(train_annotations_files))
```

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
fig=plt.figure(figsize=(20, 20))
columns = 2
rows = 5
i=1
img=[i for i in range (11)]
while i<11:

    img[i] = cv2.imread(train_data_files[i])
    img[i] = cv2.cvtColor(img[i], cv2.COLOR_BGR2GRAY)
    img[i] = cv2.resize(img[i], (224,224))
    fig.add_subplot(rows, columns, i)
    plt.imshow(cv2.cvtColor(img[i], cv2.COLOR_BGR2RGB))
    i=i+1;
plt.show()
    
```

```python
from sklearn.cluster import KMeans

fig=plt.figure(figsize=(20, 20))
columns = 2
rows = 5
i=1
cluster_pic=[i for i in range(11)]
while i<11:

    pic_n = img[i].reshape(img[i].shape[0]* img[i].shape[1])
    pic_n = pic_n.reshape(-1,1)
    kmeans = KMeans(n_clusters=2, random_state=0).fit(pic_n)
    pic2show = kmeans.cluster_centers_[kmeans.labels_]
    cluster_pic[i] = pic2show.reshape(img[i].shape[0], img[i].shape[1])
    fig.add_subplot(rows, columns, i)
    plt.imshow(cluster_pic[i])
  
    i+=1
plt.show()
    
```
